```python
def outliers_remove(df, y):
    q75, q25 = np.percentile(df[y], [95, 5])
    iqr = q75 - q25
    max = q75 + (1.5 * iqr)
    min = q25 - (1.5 * iqr)
    df.loc[df[y] < min, y] = df[y].mean()
    df.loc[df[y] > max, y] = df[y].mean()
    return df
```

```python
def bestalpha(x_train, x_test, y_train, y_test, modeltype):
    bestRsq = 0.0  # To contain the best R^2.
    bestLambda = 0.0  # To contain the best lambda.
    for aLambda in range(1, 100):
        if (modeltype == "ridge"):
            model = Ridge(aLambda)  # Train with the training set.
        elif (modeltype == "lasso"):
            model = Lasso(aLambda)
        model.fit(x_train, y_train)  # Calculate R^2 with the testing set.
        current_score = model.score(x_test, y_test)
        if (bestRsq < current_score):  # If R^2 is better, replace bestRsq.
            bestRsq = current_score
            bestLambda = aLambda
    return bestRsq, bestLambda
```

```python
X = df.drop(columns=["id","price", "date", "day"], axis=1)
y = df[["price"]]
```

```python
l = LinearRegression()
ll = Lasso(bestalpha(X_train, X_test, Y_train, Y_test,"lasso")[1])
lr = Ridge(bestalpha(X_train, X_test, Y_train, Y_test,"ridge")[1])

l.fit(X_train, Y_train)
ll.fit(X_train, Y_train)
lr.fit(X_train, Y_train)
```

```python
print("Linear accuracy:",l.score(X_test, Y_test))
print("Lasso accuracy:", ll.score(X_test, Y_test))
print("Ridge accuracy:", lr.score(X_test, Y_test))
```
